```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("hack/raw100-16-inclusive.tsv",sep='\t',index_col=0)
df100 = df.groupby(df.index).sum()
df100.to_excel('df100.xls', sheet_name='Sheet1', index=False, engine='xlsxwriter')
df = pd.read_csv("hack/raw9-16-inclusive.tsv",sep='\t',index_col=0)
df9 = df.groupby(df.index).sum()
df9.to_excel('df9.xls', sheet_name='Sheet1', index=False, engine='xlsxwriter')
```

```python
diff = df100.sub(df9, fill_value=0)
```

```python
sdiff = diff.sort_values(by=["Hardware Event Count:CPU_CLK_UNHALTED.THREAD"])
```

```python
cdf = sdiff.corr()
sdiff['Callpath'] = sdiff.index
sdiff.to_excel('diff.xls', sheet_name='Sheet1', index=False, engine='xlsxwriter')
```
